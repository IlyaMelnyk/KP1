```python
dfm = df
# Outliers
торг_площадь = []
for торг_площ in df['торг_площ']:
    if торг_площ > (df['торг_площ'].mean() + 3*df['торг_площ'].std()):
        торг_площ = (dfm['торг_площ'].mean() + 3*df['торг_площ'].std())
    elif торг_площ < (df['торг_площ'].mean() - 3*df['торг_площ'].std()):
        торг_площ = (dfm['торг_площ'].mean() - 3*df['торг_площ'].std())
    торг_площадь.append(торг_площ)
df['торг_площ'] = торг_площадь

кол_касс = []
for колич_касс in df['колич_касс']:
    if колич_касс > (df['колич_касс'].mean() + 3*df['колич_касс'].std()):
        колич_касс = (dfm['колич_касс'].mean() + 3*df['колич_касс'].std())
    elif колич_касс < (df['колич_касс'].mean() - 3*df['колич_касс'].std()):
        колич_касс = (dfm['колич_касс'].mean() - 3*df['колич_касс'].std())
    кол_касс.append(колич_касс)
df['колич_касс'] = кол_касс

доход_населен = []
for доход_насел in df['доход_насел']:
    if доход_насел > (df['доход_насел'].mean() + 3*df['доход_насел'].std()):
        доход_насел = (dfm['доход_насел'].mean() + 3*df['доход_насел'].std())
    elif доход_насел < (df['доход_насел'].mean() - 3*df['доход_насел'].std()):
        доход_насел = (dfm['доход_насел'].mean() - 3*df['доход_насел'].std())
    доход_населен.append(доход_насел)
df['доход_насел'] = доход_населен
```

```python
def rf_max_leaf_nodes(X_train, X_test, y_train, y_test, n):
    mse_train = []
    mse_test = []
    for i in n:
        rf = RandomForestRegressor(n_estimators=6, max_leaf_nodes =i, random_state=10).fit(X_train, y_train)
        mse_train.append(mean_squared_error(y_train, rf.predict(X_train)))
        mse_test.append(mean_squared_error(y_test, rf.predict(X_test)))
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(n, mse_train, alpha=0.5, color='blue', label='train')
    ax.plot(n, mse_test, alpha=0.5, color='red', label='test')
    ax.set_ylabel("MSE")
    ax.set_xlabel("max_leaf_nodes")
    
# The optimal number of max_leaf_nodes
rf_max_leaf_nodes(X_train[f], X_test[f], y_train, y_test, [5, 10, 20, 30])
```
